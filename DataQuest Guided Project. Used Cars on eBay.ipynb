We see that the column names use camelcase. Python's preferred is snake_case, so here is how we correct the column names.

```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'odometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'num_photos', 'postal_code',
       'last_seen']
autos.head()
```

'Price' and 'Odometer' columns have numeric values stored as text. We need to convert these columns to a numeric type.

```python
autos["price"] = autos["price"].str.replace('$', '')
autos["price"] = autos["price"].str.replace(',', '')
autos["price"] = autos["price"].astype(int)
```

```python
autos["odometer"] = autos["odometer"].str.replace('km', '')
autos["odometer"] = autos["odometer"].str.replace(',', '')
autos["odometer"] = autos["odometer"].astype(int)
```

```python
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
```

We can also remove columns that contain only one value and hence are not interesting for our analysis.

```python
autos = autos.drop(["num_photos", "offer_type", "seller"], axis=1)
```

```python
autos.describe(include = 'all')
```

We noticed there were some outliers in the Price column and invalid entries in the Registration Year column. Removed those.

```python
autos = autos[autos["price"].between(1,351000)]
autos["price"].describe()
```

```python
autos = autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).head(10)
```

# Part II. Data Cleaning and Data Analysis Tasks

1) To continue with our data cleaning tasks, we'll identify categorical data that uses German words, translate them and map the values to English counterparts.

```python
autos.head()
```

The following columns have German words that need to be translated: Vehicle Type, Gearbox, Model, Fuel Type, and Unrepaired Damage. We'll create a dictionary and replace German words with their English counterparts.

```python
german_eng_dict = {'andere': 'other', 'kleinwagen': 'compact car', 'kombi': 'station wagon', 'automatik': 'automatic', 'manuell': 'manual', 'benzin': 'gasoline', 'elektro': 'electro', 'ja': 'yes', 'nein': 'no'}
autos["vehicle_type"] = autos["vehicle_type"].map(german_eng_dict).fillna(autos["vehicle_type"])
autos["gearbox"] = autos["gearbox"].map(german_eng_dict).fillna(autos["gearbox"])
autos["model"] = autos["model"].map(german_eng_dict).fillna(autos["model"])
autos["fuel_type"] = autos["fuel_type"].map(german_eng_dict).fillna(autos["fuel_type"])
autos["unrepaired_damage"] = autos["unrepaired_damage"].map(german_eng_dict).fillna(autos["unrepaired_damage"])
```

```python
autos.head()
```

2) Now let's look at the columns containing dates. We'll convert the timestamps to uniform numeric data, for example "2016-03-21 13:38:56" becomes the integer "20160321".

```python
autos["date_crawled"] = autos["date_crawled"].str.split(n=1, expand=True).iloc[:, 0]
autos["date_crawled"] = autos["date_crawled"].str.replace('-', '')
autos["date_crawled"] = autos["date_crawled"].astype(int)
```

```python
autos["ad_created"] = autos["ad_created"].str.split(n=1, expand=True).iloc[:, 0]
autos["ad_created"] = autos["ad_created"].str.replace('-', '')
autos["ad_created"] = autos["ad_created"].astype(int)
```

```python
autos["last_seen"] = autos["last_seen"].str.split(n=1, expand=True).iloc[:, 0]
autos["last_seen"] = autos["last_seen"].str.replace('-', '')
autos["last_seen"] = autos["last_seen"].astype(int)
```

### Aggregation Analysis Technique

Aggregation is a useful technique to uncover patterns in the data.  
Here are the steps:  
    -identify the unique values we want to aggregate by  
    -create an empty dictionary to store our aggregate data  
    -loop over the unique data, and for each 1) subset the dataframe by the unique values, 2) calculate the mean, 3) assign the value/mean to the dictionary as key/value pairs

### Price Patterns Based on Mileage

Let's split the "odometer_km" column into groups, and use aggregation analysis technique to see if average prices follow any patterns based on the mileage.

```python
autos["odometer_km"].value_counts(normalize=True)
```

The values in this column are rounded, probably because sellers only had pre-set options for this field. As one would expect, there are more high mileage than low mileage vehicles.

```python
mileage_groups = autos["odometer_km"].value_counts(normalize=True).index
mileage_groups
```

```python
mileage_mean_prices = {}
for group in mileage_groups:
    mileage_group = autos[autos["odometer_km"] == group]
    mean_price = mileage_group["price"].mean()
    mileage_mean_prices[group] = int(mean_price)
mileage_mean_prices
```

There is a pattern there: the more miles the car has, the cheaper it is. All else equal, sellers can charge a premium if a car has low mileage.

###  Price Patterns Based on Damage History

Let's use aggregation analysis technique to see how much cheaper damaged cars are compared to non-damaged ones.

```python
damage_yes_no = autos["unrepaired_damage"].value_counts().index
damage_yes_no
```

```python
damage_price_dict = {}
for n in damage_yes_no:
    damage = autos[autos["unrepaired_damage"] == n]
    mean_price = damage["price"].mean()
    damage_price_dict[n] = int(mean_price)
damage_price_dict
```

Damaged cars are on average 3 times cheaper than non-damaged ones.
